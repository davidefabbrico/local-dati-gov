# Vaccinazione Italiana

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("mode.chained_assignment", None)
import sys
import geopandas as gpd
```

OpenData: https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/somministrazioni-vaccini-latest.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/somministrazioni-vaccini-latest.csv")
```

```python
# Correzione Dataframe 90+
for i in range(len(df)):
    df["fascia_anagrafica"][i] = df["fascia_anagrafica"][i].replace(" ", "")
```

```python
for i in range(len(df)):
    df['data_somministrazione'][i] = pd.to_datetime(df['data_somministrazione'][i]) # Dtata
```

```python
df_group1 = df.groupby("data_somministrazione").sum()
```

# Sesso / Categorie vaccinate per data di somministrazione

```python
df_group1.plot(figsize = (30, 10),  grid = False, colormap = "Accent", kind = "bar",
              title = "Categorie Vaccinazioni aggiornamento del 10/01/2021", sharex = True, subplots = True)
df_group1.plot(figsize = (30, 10),  grid = False, colormap = "Accent", kind = "bar",
              title = "Categorie Vaccinazioni aggiornamento del 10/01/2021", sharex = True)
df_group1.plot(figsize = (30, 10),  grid = False, colormap = "Accent", kind = "line",
              title = "Categorie Vaccinazioni aggiornamento del 10/01/2021", sharex = False)
plt.show()
```

# Raggruppamento per Regione

```python
df_area = df.groupby("area").sum()
df_area.plot(figsize = (30, 10),  grid = False, colormap = "tab20c", kind = "bar",
              title = "Regioni Vaccinazioni aggiornamento del 10/01/2021")
plt.show()
```

# Totali Regione

```python
serie_area_tot = df_area.sum(axis = 1)
plt.figure(figsize = (20,10))
plt.bar(serie_area_tot.index, serie_area_tot.values, color = "#fa9040", edgecolor = "black")
plt.show()
```

# Fascia Anagrafica

```python
df_eta = df.groupby("fascia_anagrafica").sum()
df_eta.plot(figsize = (30, 10),  grid = False, colormap = "Accent", kind = "bar",
              title = "Categorie Vaccinazioni aggiornamento del 10/01/2021", sharex = False)
plt.show()
```
